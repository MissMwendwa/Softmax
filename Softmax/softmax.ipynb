```python
import numpy as np
```

```python
def softmax(vector):
    exponential = np.exp(vector)
    probabilities = exponential/np.sum(exponential)
    return probabilities

vector_num = np.array([1.0, 2.0, 3.0, 4.0])
probabilities = softmax(vector_num)
print(probabilities)
```

```python
x = vector_num
y = probabilities
```

```python
import matplotlib.pyplot as plt

figure, axis = plt.subplots(figsize=(7,7))
plt.plot(x, y)
plt.xlabel("Vector Numbers")
plt.ylabel("Probabilities")
plt.show
```
